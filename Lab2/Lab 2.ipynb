```python
%matplotlib notebook
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

from sympy.interactive import printing
printing.init_printing(use_latex=True)
```

# Biblioteka Sympy

## zad 1

```python
x, y = sym.symbols('x y')
polynomial = x**3 + 3*x**2 - 10*x + 24
polynomial

x_axis = np.arange(-5, 5, 0.1)
y_axis = [polynomial.evalf(subs={'x': x}) for x in x_axis]
plt.plot(x_axis, y_axis)
plt.xlim(-5, 5)

#pierwiastki wielomianu
sym.solve(polynomial, x)
```

## zad 2 i 3

```python
poly1 = sym.Eq(x**2 + 3*y, 10)
poly2 = sym.Eq(4*x - y**2, -2)
result = sym.solve([poly1, poly2], [x,y])
result
```

## zad 4

```python
for s in result:
    print(f' S0 -> {s[0]},\n S1 -> {s[1].evalf()} \n')
    print('----')
```
